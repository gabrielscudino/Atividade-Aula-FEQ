### Atividade Reator CSTR

```python
from dataclasses import dataclass
@dataclass
class ParametrosCSTR:
   """
   Vazão : Vazão da corrente em L/min
   Volume  : Volume L
   Velocidade Específica : 1/min

   """
   vazao : float
   volume : float
   velocidade_especifica : float
```

```python
def modelo_cstr(Cain,params):
  F= params.vazao
  V = params.volume
  k = params.velocidade_especifica
  Ca = ((F)/(F+(V*k)))*Cain
  return Ca
```

```python
from re import X
def simula_cstr(num_estagios = 3):
  Ca0 = 1.0
  params = ParametrosCSTR(vazao=50.0,volume=100.0,velocidade_especifica=0.5)
  Ca = modelo_cstr(Ca0,params)
  Calist = []
  Xlist = []
  for i in range(num_estagios):
    Ca = modelo_cstr(Ca,params)
    X = ((Ca0-Ca)/Ca0)
    Calist.append(Ca)
    Xlist.append(X)
    return Calist,Xlist
```
