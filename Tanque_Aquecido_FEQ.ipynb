```python
from dataclasses import dataclass
@dataclass
class ParametrosAquecimento:
   """
   T = temperatura do líquido (°C)
   Q = potência de aquecimento (W)
   U = coeficiente global de troca térmica (W/(m⋅°C))
   A = área de troca térmica (m2)
   Tamb = temperatura ambiente (°C)
   m = massa de líquido (kg)
   cp = calor específico (J/(kg ⋅°C))

   """
   T: float=25.0
   Q: float=5000.0
   U: float=150.0
   A: float=5.0
   Tamb: float=25.0
   m: float=50.0
   Cp: float=1670.0
```

```python
def modelo_aquecimento(t, T, params):
  T= params.T
  Q= params.Q
  U = params.U
  A= params.A
  Tamb= params.Tamb
  m = params.m
  Cp = params.Cp
  dTdt = ((Q - U * (T - Tamb) * A) / (m * Cp))
  return dTdt
```

```python
import numpy as np
from scipy.integrate import solve_ivp
def simula_aquecimento(tempo, params):
  t_span = (0, tempo)
  Cond_inicial = 25
  t_eval = (np.linspace(0, tempo, 100))
  sol = solve_ivp(modelo_aquecimento, t_span, Cond_inicial, t_eval=t_eval, args =params,method='RK45')
  return sol.t, sol.y[0]
```
